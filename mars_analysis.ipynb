```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('mars-weather.csv')
```

```python
# Calculate the correlation matrix
corr_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Select columns for histogram
numeric_columns = ['sol', 'ls', 'min_temp', 'max_temp', 'pressure', 'atmo_opacity']

# Number of rows and columns for subplots
n_rows = 4
n_cols = 3

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 20))

# Flatten the axes
axes = axes.flatten()

# Loop through numeric_columns and plot histograms
for i, column in enumerate(numeric_columns):   
    axes[i].hist(df[column], bins=20, edgecolor='black')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frequency')
    axes[i].set_title(f'Histogram of {column}')

# Line plot for 'terrestrial_date' against 'max_temp'
axes[i+1].plot(df['terrestrial_date'], df['max_temp'], color='blue')
axes[i+1].set_xlabel('Date')
axes[i+1].set_ylabel('Max Temp')
axes[i+1].set_title('Max Temp over Time')

# Remove extra subplots
for j in range(i+2, n_rows*n_cols):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
